```python
import pandas as pd
data=pd.read_csv("btissue.csv")
data.head()
```

```python
import numpy as np
np.shape(data)
```

```python
from sklearn.tree import DecisionTreeClassifier
predictors=data.iloc[:,0:9]
predictors.head()
```

```python
target=data.iloc[:,9]
target.head()
```

```python
from sklearn.model_selection import train_test_split

predictors_train, predictors_test, target_train, target_test = train_test_split(
    predictors, target, 
    test_size=0.3,       # 30% of data goes to the test set
    random_state=123     # ensures reproducibility
)
```

```python
predictors_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree_entropy = DecisionTreeClassifier(
    criterion="entropy",
    random_state=100,
    max_depth=3,
    min_samples_leaf=5
)
dtree_entropy.fit(predictors_train, target_train)

prediction = dtree_entropy.predict(predictors_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(target_test, prediction, normalize=True)
```

```python
from sklearn.metrics import confusion_matrix
target_test
```

```python
confusion_matrix(target_test,prediction)
```
