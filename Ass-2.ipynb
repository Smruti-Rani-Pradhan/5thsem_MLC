```python
import matplotlib.pyplot as plt

# Data
discount = [10, 20, 30, 40, 50]
sales = [40000, 45000, 48000, 50000, 100000]

# Plotting
plt.scatter(discount, sales, color='blue')
plt.title('Sales vs Discount Offered')
plt.xlabel('Discount (%)')
plt.ylabel('Sales (in Rs)')
plt.grid(True)
plt.show()
```
