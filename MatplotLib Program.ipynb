```python
import matplotlib.pyplot as plt
date=["25/12","26/12","27/12"]
temp=[8.5,10.5,6.8]
plt.plot(date,temp)
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("Date wise Temperature")
plt.grid(True)
plt.yticks(temp)
plt.show()

```

```python
import matplotlib.pyplot as plt
X=range(-15,16)
Y=[x**2 for x in X]
plt.plot(X,Y,'r')
plt.xlabel('X axis points')
plt.ylabel('Y axis points')
plt.title('Plotting y=x^2')
plt.show()
```

```python
import numpy as np
np.random.seed(42)
X=np.random.rand(50)
Y=np.random.rand(50)
plt.scatter(X,Y,color='red',marker='o',s=100,edgecolors='black',alpha=0.5,label='Scatter Plot')
plt.legend(loc=9)
plt.show()

```
